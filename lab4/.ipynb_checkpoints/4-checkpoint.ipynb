```python
# check normality, for example - 
# The shapiro() SciPy function will calculate the Shapiro-Wilk on a given dataset

from scipy.stats import shapiro
# can use a for loop
stat1, p1 = shapiro(list(df_numeric['Customer Lifetime Value']))
stat2, p2 = shapiro(list(df_numeric['Income']))
stat3, p3 = shapiro(list(df_numeric['Months Since Policy Inception']))

print('Statistics=%.3f, p=%.3f' % (stat1, p1))
print('Statistics=%.3f, p=%.3f' % (stat2, p2))
print('Statistics=%.3f, p=%.3f' % (stat3, p3))
```

```python
# dummy
cat = df.select_dtypes(include = np.object)
cat = cat.drop(['Customer', 'State','Effective To Date', 'Policy Type', 'Location Code'], axis=1)
categorical = pd.get_dummies(cat,drop_first=True)
categorical.head()
```

```python
df_num = df.select_dtypes(include=['int64', 'float64'])
df_num
```

```python
df_num = df_num.drop(['Months Since Last Claim', 'Months Since Policy Inception','Number of Open Complaints','Number of Policies'], axis=1)
df_num
```

```python
sns.displot(df_num['Customer Lifetime Value'])
plt.show()

sns.displot(df_num['Income'])
plt.show()

sns.displot(df_num['Monthly Premium Auto'])
plt.show()
# high skewness
```

```python
# Normalizing data
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler

df_num1 = df_num.drop(['Total Claim Amount'], axis=1)
df_num2 = df_num.drop(['Total Claim Amount'], axis=1)

transformer = Normalizer().fit(df_num1)
x_normalized = transformer.transform(df_num1)
x_normalized = pd.DataFrame(x_normalized)
x_normalized

transformer = StandardScaler().fit(df_num1)
x_standarized = transformer.transform(df_num1)
x_standarized = pd.DataFrame(x_standarized)
```

```python
sns.displot(x_standarized.iloc[:, 0])
plt.show()
```

```python
sns.displot(x_standarized.iloc[:, 1])
plt.show()
```

```python
y = df['Total Claim Amount']
y
```

```python
X = np.concatenate((x_normalized, categorical), axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train, y_train)

reg.score(X_test, y_test)
```
